## For masking certain region using mask

 the mask matches the footage dimension (1920x1080) , choose frame according to the footage

```python
import cv2
import cvzone
import math

# Load the mask image
mask = cv2.imread("mask.png", cv2.IMREAD_GRAYSCALE)

# Create VideoCapture object
cap = cv2.VideoCapture("../Car counters/stock_footage.mp4")  
cap.set(3, 1280)
cap.set(4, 720)

while True:
    # Read frame from webcam
    ret, frame = cap.read()
    if not ret:
        break  # Break the loop if there are no more frames
    
    # Apply the mask to the frame
    masked_frame = cv2.bitwise_and(frame, frame, mask=mask)

    # Process results
    results = model(masked_frame, stream=True)
    for r in results:
        boxes = r.boxes
        for box in boxes:
            # for bounding boxf
            x1,y1,x2,y2=box.xyxy[0]
            x1,y1,x2,y2=int(x1),int(y1),int(x2),int(y2)
            # cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,255),3)
            w,h=x2-x1,y2-y1
            
            # for confidence
            conf=math.ceil((box.conf[0]*100))/100
            print(conf)
            # class names
            cls=int(box.cls[0]) # since the value was floating we have converted into integer value

            if coco_classes[cls] == "car" or coco_classes[cls] == "Truck" or coco_classes[cls] == "Bus" or coco_classes[cls] == "Motorcycle"\
            and cnf > 0.3:
                cvzone.putTextRect(frame,f'{coco_classes[cls]} {conf}',(max(0,x1),max(35,y1)),offset=5)
            # cvzone.putTextRect(frame,f'{conf}',(max(0,x1),max(35,y1)))
                cvzone.cornerRect(frame,(x1,y1,w,h),l=10,colorR=(255, 0, 255),
    colorC=(0, 255, 0))

    # Display the frame with overlay
    cv2.imshow('Footage', frame)
    
    # Wait for key press
    if cv2.waitKey(1) & 0xFF == ord('q'):  
        break  # Press 'q' to exit

# Release resources
cap.release()
cv2.destroyAllWindows()
```
